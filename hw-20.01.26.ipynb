```python
# Take input from the user
N = int(input("Enter a positive integer: "))

# Using for loop to print numbers from 1 to N
for i in range(1, N + 1):
    print(i, end=" ")  
```

```python
# Take input from the user
N = int(input("Enter a positive integer: "))

# Initialize counter
i = 1

# Using while loop
while i <= N:
    print(i, end=" ")
    i += 1  # increment counter
```

```python
# Input from user
N = int(input("Enter a positive number: "))
while N > 0:
    digit = N % 10      
    print(digit)        
    N = N // 10          
```

```python
# Input from user
n = int(input("Enter a positive number: "))

print("Factors of", n, "are:")
for i in range(1, n + 1):
    if n % i == 0:   
        print(i, end=" ") 
```

```python
# Input from user
n = int(input("Enter a positive number: "))

count = 0 
for i in range(1, n + 1):
    if n % i == 0:   
        count += 1   

print("Number of factors:", count)
```

```python
num = int(input("Enter a positive number: "))

if num > 1 and all(num % i != 0 for i in range(2, num)):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```
